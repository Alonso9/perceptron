```python
import numpy as np
import matplotlib.pyplot as plt

class perceptron:

    def __init__(self, num_inputs, lr, epochs, pesos=None):
      if pesos:
        self.weights = pesos
      else:
        self.weights = np.random.rand(num_inputs+1) # se crean pesos de manera random
      self.lr  = lr # Learning rate
      self.epochs = epochs # Número de épocas

    def _init_(self, number_of_epoch):
        self.input = np.array([0, 0])
        self.weights1 = np.random.rand(4, 2)
        self.ideal = 0
        self.weights2 = np.random.rand(1, 4)
        self.output = np.zeros(1)
        self.number_of_epoch = number_of_epoch

    def act_fn(self, x, function='step'): # Función de activación
      if function == 'step':
        return np.where(x>1,  1, 0)
      if function == 'sigmoid':
        return 1/(1+np.exp(-x))
      if function == 'relu':
        return np.where(x>0,x,0)
      if function == 'tanh':
        return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    
    def act_fn(self, x, derv=False, function='sigmoid'): #Activation fuction
      if function == 'step':
        return np.where(x>1, 1, 0)
      if function == 'sigmoid':
        if derv:
          return x * (1 - x)
        return 1/(1+np.exp(-x))
      if function == 'relu':
        return np.where(x>0,x,0)
      if function == 'tanh':
        return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

    def predict(self, inputs):
      return self.act_fn(np.dot(inputs, self.weights[1:]) + self.weights[0])

    def feedforward(self):
        self.layer = self.act_fn(np.dot(self.input, self.weights1.T))
        self.output = self.act_fn(np.dot(self.layer, self.weights2.T))
    
    def backprop(self):
        # application of the chain rule to find derivative of the loss function with respect to weights2 and weights1
        delta2 = 2 * (self.ideal - self.output) * self.act_fn(self.output, derv=True)
        delta1 = np.dot(self.weights2.T, delta2) * self.act_fn(self.layer, derv=True)
        d_weights2 = np.multiply(delta2, self.layer.T)
        d_weights1 = np.dot(delta1.reshape(4, 1), self.input.reshape(1, 2))
        # update the weights with the derivative (slope) of the loss function
        self.weights2 += d_weights2
        self.weights1 += d_weights1


    def train(self, in_train, clases, verbose=False ):
      errores = []
      for _ in range(self.epochs):
        e = 0
        for entrada, clase in zip(in_train, clases):
          prediccion = self.predict(entrada)
          error = clase -prediccion
          self.weights[0] += self.lr * error # actualización de pesos
          self.weights[1:] += self.lr * error * entrada

          e += np.abs(error)
        errores.append(e)

      if verbose:
        plt.figure(), plt.plot(errores), plt.title('errores'), plt.show()

    def get_weights(self):
      return self.weights    

x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], int)
y = np.array([[0], [1], [1], [0]], int)

neurona = perceptron(2,0.2,10)
print(neurona.get_weights())

for entradas, label in zip(x,y):
  prediccion = neurona.predict(entradas)
  print('entradas:', entradas, 'salida real', label, 'Salida de la red', prediccion)

neurona.train(x,y, True)

for entradas, label in zip(x,y):
  prediccion = neurona.predict(entradas)
  print('entradas:', entradas, 'salida real', label, 'Salida de la red', prediccion)

print(neurona.get_weights())
```
